**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#calculate the average house price for each year and save it to a new dataframe 
dfg = df.groupby(['year','borough']).mean()
dfg = dfg.reset_index()
dfg[dfg['year']==1996]
```

```python
idx = dfg.groupby(['year'])['house_price'].transform(max) == dfg['house_price']
idx
```

```python
a = dfg.groupby("year")["house_price"].idxmax()
```

```python
result = dfg.iloc[a,0:2]
result
```

```python
def create_price_ratio(d):
    y1998 = float(d['house_price'][d['year']==1998])
    y2018 = float(d['house_price'][d['year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
#  Test out the function by calling it with the following argument:
# dfg[dfg['London_Borough']=='Barking & Dagenham']
create_price_ratio(dfg[dfg['borough']=='Barking & Dagenham'])
```

```python
# for bo, it takes a subset of the dataset for each borough(data from 1995 to 2021), when running the function, the create_price_ratio is only running on that subset. what is how the loop works 
final = {}
for i in dfg['borough'].unique(): 
    bo = dfg[dfg['borough']==i]
    final[i] = create_price_ratio(bo)  
final
```

```python
#make the list to dataframe
df_final_d = pd.DataFrame(final)
df_final_d
```

```python
#transpose and rename the columns 
df_final = df_final_d.T
df_final = df_final.reset_index()
df_final = df_final.rename(columns={'index':'borough',0:'ratio_2018_1998'})
df_final
```

```python
#which boroughs of London have seen the greatest house price increase, on average, over the past two decades?
#sort the dataframe 
df_final = df_final.sort_values('ratio_2018_1998', ascending=False)
df_final 
```

```python
# Summary statistics 
#avg = df_final['ratio_2018_1998'].mean()
print('Mean of housing price increase (1998-2018) in London is:', df_final['ratio_2018_1998'].mean()) 
print('Median of housing price increase in London (1998-2018) is:', df_final['ratio_2018_1998'].median()) 
print('SD of housing price increase in London (1998-2018) is:', df_final['ratio_2018_1998'].std()) 
```

```python
# the top 15 
london_15 = df_final.head(15)
london_15
```

```python
london_bar = london_15.set_index('borough')
london_bar.plot(kind='bar',title="Top 15 boroughs' house price increases in London")
plt.show()
```
